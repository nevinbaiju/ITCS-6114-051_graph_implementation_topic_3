```python
class node:
    def __init__(self, label):
        """
        Parameter: a string indicating the label of the new node
        """
        self.label = label
        self.in_degree = 0
        self.out_degree = 0
        
    def in_degree(self):
        """
        Parameter: none
        Return value: integer representing the in-degree of 
        this node
        """
            
        return self.in_degree
    
    
    def update_in_degree(self, new_edge=0):
        """
        Parameter: none
        Return value: integer representing the in-degree of 
        this node
        """
        if new_edge == 1:
            self.in_degree += 1
        elif new_edge == -1:
            self.in_degree -= 1
            
        return self.in_degree
        
    
    def out_degree(self):
        """
        Parameter: none
        Return value: integer representing the out-degree of 
        this node
        """
        
        return self.out_degree
        
    
    def update_out_degree(self, new_edge=0):
        """
        Parameter: none
        Return value: integer representing the out-degree of 
        this node
        """
        if new_edge == 1:
            self.out_degree += 1
        elif new_edge == -1:
            self.out_degree -= 1
        return self.out_degree
    
    def __str__(self):
        """
        Parameter: none
        Return value: string label of the node
        """
        return self.label
```
